```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Output/merge_stud_batla.csv") 
```

```python
df["wage_std"] = (df["wage_per_month"] - df["wage_per_month"].mean()) / df["wage_per_month"].std()
```

```python
piv = df.groupby("intervention_index")["wage_per_month"].agg(["mean", "std"]).reset_index()
```

```python
piv = piv.loc[(piv["intervention_index"] > -36)]
piv = piv.loc[(piv["intervention_index"] < 36)]
```

```python
sns.lineplot(x="intervention_index", y="mean", data=piv)
plt.axvline(x=0, color="red", linestyle="--")
```
